```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/gdrive/MyDrive/DLNLP/online_shoppers_intention.csv")
```

```python
data.isnull().sum()
```

```python
data.drop(columns = "Month", axis = 1 , inplace = True)
```

```python
ss = StandardScaler()
ss.fit(data)
X = ss.transform(data)
```

```python
y = data.pop ('Revenue')
```

```python
print(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.25)
display(X_train.shape)
display(X_test.shape)
```

```python
model = tf.keras.Sequential()
```

```python
model.add( tf.keras.layers.Input(shape = (13,) ))
```

```python
model.add(tf.keras.layers.Dense(20, activation = 'relu'))
```

```python
model.add(tf.keras.layers.Dense(25, activation = 'relu'))
model.add(tf.keras.layers.Dense(20, activation = 'relu'))
model.add(tf.keras.layers.Dense(15, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss = 'mae',
    optimizer='adam')
```

```python
run = model.fit(X_train,y_train,
                    epochs = 200,
                    validation_data = (X_test, y_test)
                   )
```

```python
model.evaluate(X_test,y_test)
```
